```python
students = ("Clueless", "Eliminate 1/4", "Eliminate 2/4", "Know It All")
probs = (
    (0.25, 0.25),
    (32.0/96, 35.0/96),
    (4.0/8, 5.0/8),
    (1., 1.)
)

n = 100
plt.figure(figsize=(13, 10))
X = [i for i in range(n+1)]
for stu in range(4):
    distributions = [i for i in map(lambda q: [math.comb(n, s) * (q**s * (1-q)**(n-s)) for s in range(n+1)], probs[stu]) ]
    plt.subplot(2, 2, stu+1)
    plt.plot(X, distributions[0], '.-', label="Dumbproach ")
    plt.plot(X, distributions[1], '.--', label="Monteproach ")
    plt.title(students[stu])
    plt.xlabel("Score out of 100")
    plt.ylabel("Probabality")
    plt.legend()
plt.suptitle(f"Probabality distributions for students on 4 choice {n} question test")
None
```
